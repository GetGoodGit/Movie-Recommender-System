```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.sample(2)
```

```python
## droping null values
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
# check all available genres
id_set = set()
def get_genres(l):
    for i in range(len(l)):
        id_set.add(l[i]['name'])

lst = movies['genres'].apply( lambda x: json.loads(x))       
lst.apply(get_genres)
id_set
```

```python
# genres column data extraction
def row_genres(row):
    lst = []
    for i in row:
        lst.append(i['name'])
    return lst

lst = movies['genres'].apply( lambda x: json.loads(x))
movies['genres'] = lst.apply(row_genres)
movies['genres']
```

```python
# keywords column
movies['keywords'][0]
```

```python
# keywords column data extraction
def row_keywords(row):
    lst = []
    for i in row:
        lst.append(i['name'])
    return lst

lst = movies['keywords'].apply( lambda x: json.loads(x))
movies['keywords'] = lst.apply(row_keywords)
movies['keywords']
```

```python
# cast column data extraction
def row_cast(row):
    lst = []
    count=0
    for i in row:
        if count == 10:
            break
        lst.append(i['name'])
        count+=1
    return lst

lst = movies['cast'].apply( lambda x: json.loads(x))
movies['cast'] = lst.apply(row_cast)
movies['cast']
```

```python
movies['crew'][0]
```

```python
def row_cast(row):
    val=[]
    for i in row:
        if i['job']=='Director':
            val.append(i['name'])
    return val

lst = movies['crew'].apply( lambda x: json.loads(x))
movies['crew'] = lst.apply(row_cast)
movies['crew']
```

```python
movies.sample(5)
```

```python
# overview
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview']
```

```python
# removing space
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
## creating tag
```

```python
movies['tag'] = movies['overview']+movies['genres']+movies['cast']+movies['crew']+movies['keywords']
movies['tag'][0]
```

```python
new_df = movies[['movie_id','title','tag']]
new_df
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
# stemming - to get single word for similar word like(working, work,worked)
# !pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

```python
# using countVectorizer - convert a collection of text documents to a maxtrix of token counts
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 10000, stop_words='english')
```

```python
# returns sparse matrix so convert to np array
vectors = cv.fit_transform(new_df['tag']).toarray()
```

```python
np.set_printoptions(threshold=np.inf) # to see all elements in array
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_df.iloc[i[0]].title)
        

recommend('Avatar')
```
